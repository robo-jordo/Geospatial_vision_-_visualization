## Point matching

````python
import numpy as np
from haversine import haversine
def dist(p1,p2,p3):
    return np.linalg.norm(np.cross(p2-p1, p1-p3))/np.linalg.norm(p2-p1)

def closest_calc(probe_point):
    print('linked probe')
    temp = LinkData[['LinkID', 'ShapeInfo']]
    temp['dist'] = temp['ShapeInfo'].apply(lambda x: min(haversine(probe_point, [float(j) for j in i.split('/')[:2]]) for i in x.split('|')))
    return temp.iloc[temp['dist'].idxmin()]['LinkID']
````

Get data into a format where the data is grouped by link ID and then ordered by linkID then probeID then by time.

````python
grad_df = pd.DataFrame(columns=['linkID','dist','alt_change','slopes','mean_slope'])

for label,df_group in df3:
    lengths = []
    changes = []
    slopes = []
    running = 0
    for i in range(len(df5.loc[label])-1):      
        length1 = haversine(df5.loc[[label]].iloc[i].loc['points'][0],df5.loc[[label]].iloc[i+1].loc['points'][0])*1000
        change1 = df5.loc[[label]].iloc[i+1].loc['alts'][0]-df5.loc[[label]].iloc[i].loc['alts'][0]
        slope = np.arctan(change1/length1)
        lengths.append(length1)
        changes.append(change1)
        slopes.append(slope)
    for k in range(len(slopes)):
        running = running + slopes[k]
    
    if (len(slopes)>0):
        av_slope = float(running)/len(slopes)
    else:
        av_slope = None
    temp = pd.DataFrame([[label,lengths,changes,slopes,av_slope]],columns=['linkID','dist','alt_change','slopes','mean_slope'])
    grad_df = grad_df.append(temp)
grad_df.fillna("0", inplace=True)
    

        
````
